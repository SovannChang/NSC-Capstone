```python
import pandas as pd
```

NSC

```python
nsc_games_22 = pd.read_excel("Games Data Raw/NSC Games 2022.xlsx")
nsc_games_23 = pd.read_excel("Games Data Raw/NSC Games 2023.xlsx")
```

```python
nsc_games_all = pd.concat([nsc_games_22, nsc_games_23], ignore_index=True)
nsc_games_all
```

```python
# Dropping unnecessary columns
nsc_games_all.drop(["Time", "Day", "Opponent", "Poss", "Captain", "Formation", "Referee", "Match Report", "Notes"], axis=1, inplace=True)
```

```python
# Renaming some columns to match other DFs
nsc_games_all.rename(columns={"GF": "Score", "GA": "Opponent"}, inplace=True)
nsc_games_all.head(10)
```

Titans

```python
titans_games_22 = pd.read_excel("Games Data Raw/Titans Games 2022.xlsx")
titans_games_23 = pd.read_excel("Games Data Raw/Titans Games 2023.xlsx")
```

```python
titans_games_all = pd.concat([titans_games_22, titans_games_23], ignore_index=True)
with pd.option_context('display.max_columns', 1000):
    display(titans_games_all.head())
```

```python
# Dropping unnecessary columns
titans_games_all.drop(["Week", "Day", "Unnamed: 3", "Unnamed: 4", "Rec", "OT", "Opp", "1stD", "TotYd", "PassY", "RushY", "TO", "1stD.1", "TotYd.1", "PassY.1", "RushY.1", "TO.1", "Offense", "Defense", "Sp. Tms"], axis=1, inplace=True)
```

```python
# Dropping bye weeks
titans_games_all.dropna(subset=["Tm"], inplace=True)
```

```python
# Renaming columns
titans_games_all.rename(columns={"Unnamed: 5": "Result", "Unnamed: 8": "Venue", "Tm": "Score", "Opp.1": "Opponent"}, inplace=True)
```

```python
# Fixing Venue column data
titans_games_all.fillna("Home", inplace=True)
titans_games_all.loc[titans_games_all["Venue"] == "@", "Venue"] = "Away"
```

```python
# Converting Score and Opponent to integers
titans_games_all = titans_games_all.astype({'Score': 'int32', "Opponent": "int32"})
```

```python
# Manually adding in Titans' attendance
print(titans_games_all.shape)
titans_attendance = [67805, 70683, 68649, 65781, 51836, 69459, 68467, 73475, 69240, 77506, 69441, 69879, 68409, 70240, 66634, 69292, 70050, 70024, 67438, 67919, 68951, 65287, 61011, 68008, 65969]
len(titans_attendance)
```

```python
titans_games_all["Attendance"] = titans_attendance
titans_games_all.head(10)
```

Predators

```python
preds_games_22 = pd.read_excel("Games Data Raw/Preds Games 2022.xlsx")
preds_games_23 = pd.read_excel("Games Data Raw/Preds Games 2023.xlsx")
```

```python
preds_games_all = pd.concat([preds_games_22, preds_games_23], ignore_index=True)
```

```python
# Dropping unnecessary columns
preds_games_all.drop(["GP", "Opponent", "Unnamed: 7", "W", "L", "OL", "Streak", "LOG", "Notes"], axis=1, inplace=True)
```

```python
# Renaming columns
preds_games_all.rename(columns={"Unnamed: 6": "Result", "Unnamed: 2": "Venue", "GF": "Score", "GA": "Opponent", "Att.": "Attendance"}, inplace=True)
```

```python
# Fixing Venue column data
preds_games_all.fillna("Home", inplace=True)
preds_games_all.loc[preds_games_all["Venue"] == "@", "Venue"] = "Away"
```

```python
preds_games_all.head(10)
```

Add a Stadium Percent Full column for each team, with values only during home games

```python
nsc_games_all["Percent"] = 0
nsc_games_all.loc[nsc_games_all["Venue"] == "Home", "Percent"] = nsc_games_all.Attendance / 30109

titans_games_all["Percent"] = 0
titans_games_all.loc[titans_games_all["Venue"] == "Home", "Percent"] = titans_games_all.Attendance / 69459

preds_games_all["Percent"] = 0
preds_games_all.loc[preds_games_all["Venue"] == "Home", "Percent"] = preds_games_all.Attendance / 17823
```

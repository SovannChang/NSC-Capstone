```python
import pandas as pd
```

```python
posts1 = pd.read_csv("Instagram Raw Data/Instagram Posts Aug 19 '23  - Sep 27 '23.csv")
posts2 = pd.read_csv("Instagram Raw Data/Instagram Posts Sep 5 '22 - Aug 19 '23.csv")
posts_all = pd.concat([posts1, posts2], ignore_index=True)
posts_all.drop_duplicates("timestamp").reset_index(drop=True, inplace=True)              # Dropping duplicate posts because of the overlapping time frame
posts_all
```

```python
comments1 = pd.read_csv("Instagram Raw Data/Instagram Comments Posts 1-50.csv")
comments2 = pd.read_csv("Instagram Raw Data/Instagram Comments Posts 51-100.csv")
comments3 = pd.read_csv("Instagram Raw Data/Instagram Comments Posts 101-200.csv")
comments4 = pd.read_csv("Instagram Raw Data/Instagram Comments Posts 201-300.csv")
comments5 = pd.read_csv("Instagram Raw Data/Instagram Comments Posts 301-400.csv")
comments6 = pd.read_csv("Instagram Raw Data/Instagram Comments Posts 401-500.csv")
comments7 = pd.read_csv("Instagram Raw Data/Instagram Comments Posts 501-600.csv")
comments8 = pd.read_csv("Instagram Raw Data/Instagram Comments Posts 601-700.csv")
comments9 = pd.read_csv("Instagram Raw Data/Instagram Comments Posts 701-800.csv")
comments10 = pd.read_csv("Instagram Raw Data/Instagram Comments Posts 801-884.csv")
```

```python
comments_all = pd.concat([comments1, comments2, comments3, comments4, comments5, comments6, comments7, comments8, comments9, comments10], ignore_index=True)
comments_all.drop_duplicates(["timestamp", "text"]).reset_index(drop=True, inplace=True) 
comments_all
```

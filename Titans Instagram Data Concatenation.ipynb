```python
comments1 = pd.read_csv("Instagram Raw Data/Titans Comments Posts 1-100.csv")
comments2 = pd.read_csv("Instagram Raw Data/Titans Comments Posts 101-200.csv")
comments3 = pd.read_csv("Instagram Raw Data/Titans Comments Posts 201-300.csv")
comments4 = pd.read_csv("Instagram Raw Data/Titans Comments Posts 301-400.csv")
comments5 = pd.read_csv("Instagram Raw Data/Titans Comments Posts 401-500.csv")
comments6 = pd.read_csv("Instagram Raw Data/Titans Comments Posts 501-600.csv")
comments7 = pd.read_csv("Instagram Raw Data/Titans Comments Posts 601-700.csv")
comments8 = pd.read_csv("Instagram Raw Data/Titans Comments Posts 701-800.csv")
comments9 = pd.read_csv("Instagram Raw Data/Titans Comments Posts 801-900.csv")
comments10 = pd.read_csv("Instagram Raw Data/Titans Comments Posts 901-961.csv")
```

```python
posts_df.rename(columns={"timestamp": "postTimestamp", "likesCount": "postLikesCount", "ownerId": "postOwnerId", "ownerUsername": "postOwnerUsername", "id": "postId", "url": "postUrl"}, inplace=True)
```

```python
posts_df["postId"] = posts_df["postId"].astype(float)
```

Testing out NLP Cloud

```python
import nlpcloud
from getpass import getpass
import random
import time
```

```python
NLP_CLOUD_KEY = getpass()
```

Set pn_client to do positive/negative sentiment analysis.

Set em_client to do emotional sentiment analysis.

```python
pn_client = nlpcloud.Client("distilbert-base-uncased-finetuned-sst-2-english", NLP_CLOUD_KEY, gpu=False)
em_client = nlpcloud.Client("distilbert-base-uncased-emotion", NLP_CLOUD_KEY)
```

Getting a few rows of the comments dataframe and pulling in the text of the original post

```python
comments_sample = comments_df.sample(5, random_state = 42)
```

```python
comments_sample
```

```python
sample_merged = pd.merge(comments_sample, posts_df, on="postUrl")
```

```python
sample_merged
```

```python
sample_merged.columns
```

```python
sample_merged_trim = sample_merged[["text", "caption"]]
sample_merged_trim["label"] = ""
sample_merged_trim["sentiment_score"] = 0
sample_merged_trim
```

Testing the sentiment and emotional analysis

```python
for i in range(len(sample_merged_trim)):
    sentiment = pn_client.sentiment(sample_merged_trim.loc[i, "text"])
    sample_merged_trim.at[i, "sentiment"] = sentiment["scored_labels"][0]["label"]
    sample_merged_trim.at[i, "score"] = sentiment["scored_labels"][0]["score"]
    time.sleep(30)
```

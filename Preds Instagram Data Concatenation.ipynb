```python
import pandas as pd
```

```python
posts1 = pd.read_csv("Instagram Raw Data/Preds Instagram Posts Sep 5 '23 - Oct 23 '23.csv")
posts2 = pd.read_csv("Instagram Raw Data/Preds Instagram Posts Sep 5 '22 - Sep 4 '23.csv")
posts_all = pd.concat([posts1, posts2], ignore_index=True)
posts_all.drop_duplicates("timestamp").reset_index(drop=True, inplace=True)              # Dropping duplicate posts because of the overlapping time frame
posts_all
```

```python
posts_all.drop("alt", axis=1, inplace=True)
```

```python
posts_all.to_csv("Instagram Concatenated Data/Preds Instagram Posts", index=False)
```
